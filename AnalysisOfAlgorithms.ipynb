


# Big-O Notation: Explanation, Definition, and Growth of Functions

## Formal Definition of Big-O
Big-O notation describes an upper bound on the growth of a function. Formally, we say that the run time of an algorithm that takes \( T(n) \) is \( O(f(n)) \) if there exist positive constants \( c \) and \( n_0 \) such that for all \( n $\geq$ n_0 \):

\[ T(n) $\leq$ c \* f(n) \]

This means that beyond some point \( n_0 \), \( T(n) \) grows at most as fast as \( f(n) \), up to a constant factor.



### Big-O Notation: Explanation and Growth of Functions

Big-O notation describes how an algorithm’s runtime or space requirements grow as the input size \( n \) increases. It helps in comparing the efficiency of different algorithms.

#### Common Big-O Complexities and Their Growth

Here’s a table comparing how different complexities grow with increasing input size:

| \( n \)  | \( O(1) \) | \( O($\log$ n) \) | \( O(n) \) | \( O(n $\log$ n) \) | \( O(n^2) \) | \( O(2^n) \) |
|----------|----------|-------------|---------|-------------|---------|---------|
| 1,000    | 1        | 10          | 1,000   | 10,000      | 1,000,000 | \( $\sim10$^{300} \) |
| 10,000   | 1        | 13          | 10,000  | 130,000     | 100,000,000 | Huge |
| 100,000  | 1        | 17          | 100,000 | 1,700,000   | 10 billion | Immense |

```python
import numpy as np
import matplotlib.pyplot as plt

# Input sizes
n = np.linspace(1, 100, 100)

# Define functions
O_1 = np.ones_like(n)  # O(1)
O_log_n = np.log2(n)   # O(log n)
O_n = n                # O(n)
O_n_log_n = n * np.log2(n)  # O(n log n)
O_n2 = n**2            # O(n^2)
O_2n = 2**n            # O(2^n), limited range to avoid overflow

# Plot
plt.figure(figsize=(10, 6))
plt.plot(n, O_1, label="O(1)")
plt.plot(n, O_log_n, label="O(log n)")
plt.plot(n, O_n, label="O(n)")
plt.plot(n, O_n_log_n, label="O(n log n)")
plt.plot(n, O_n2, label="O(n²)")
plt.plot(n[:20], O_2n[:20], label="O(2ⁿ)")  # Limiting to 20 values for visibility

plt.ylim(0, 5000)  # Limit y-axis for clarity
plt.xlabel("Input Size (n)")
plt.ylabel("Operations")
plt.title("Growth of Big-O Functions")
plt.legend()
plt.grid()
plt.show()
```

## When analyzing code to determine its Big-O, follow these general rules:
# Rules for Analyzing Code Complexity (Big-O) in Python

### 1. Look at Loops
- A **loop that runs _n_ times** contributes **O(n)**.
- **Nested loops** multiply their complexities (e.g., two nested loops each running _n_ times → **O(n²)**).


```python
# Code Examples
# Single loop → O(n)
def single_loop(n):
    for i in range(n):
        print(i)


# Nested loops → O(n²)
def nested_loops(n):
    for i in range(n):
        for j in range(n):
            print(i, j)


# Nested loops → O(n^3)
def nested_loops(n):
    for i in range(n-8):
        for j in range(n*n*300):
            print(i, j)
```

### 2. Consecutive Statements

Sum complexities when operations happen sequentially (and drop less dominate term).

O(n) + O(n²) → O(n²) (keep the dominant term).

```python
#Consecutive Example
def consecutive_bigo(n):
    for i in range(n):  # O(n)
        print(i)
    
    #O(n²)
    for j in range(2*n):  # O(n)
        for k in range(4*n):  # O(n)
            print(j, k)  # O(1)

# Total: O(n) + O(n²) = O(n²)
```

### 3. Ignore Constants

O(2n) → O(n)

O(100n) → O(n)

O(5) → O(1)

O(n+1000000) → O(n)

```python
def constant_operations(n):
    for i in range(100):  # Runs 100 times, but doesn't depend on n
        print(i)

# Complexity: O(1) (constant time)
```

### 4. Recursive Functions

Count the number of recursive calls and work per call.  These are harder and I will never give you one a test but you should at least be exposed to it.

Recursion Tree help visualize complexity.

```python
# Recursive function: O(2ⁿ)
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

#  this is exponentail growth because each recursive call generates two recursive calls
```

### 5. Logarithmic Complexity

O(log n) arise when the algorithm cuts the number of iterations by some fraction at each iteration.

A classic example is binary search or algorithms that use a divide and conquer approach.  

```python
# Binary search → O(log n)
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1


# 
# O(log n)
def log_example2(n):
    while n > 1:
        n = n // 2
        print(n)

# 
# O(log n)
def log_example2(n):
    i=1
    while i<n:
        i = i*5
        print(n)
```

### 6. Exponential Complexity

O(2ⁿ) means the function doubles in size with each increase in n.  Generally you can describe any exponential function as one that increases by some factor at each iteration.
